Answer the given questions with **at most a sentence**.

  >a) Why do people use validation data?<br>
**Validation set is used for hyperparameter selection and model selection. If hyperparameter tuning will be done, it will not be possible to do it on test set, because we do not want our algorithm to see test set before the test time. On the other hand, if we use training set for hyperparameter tuning, then we will fit our hyperparameters on the data where we have trained our model on, so this will make our model overfit to the training data.**

  >b) What is the difference between mean squared error and mean absolute error?  <br>
**They are mostly used in linear regression. Both metrics measure the difference between the actual and the predicted values of y. MSE takes the square of the differences to get rid of the minus results if there is, whereas MAE takes the absolute of the differences. MSE makes the errors big errors even bigger by taking the squares, on the other hand MAE is more robust and shows the real difference instead of squares.**

  >c) What is the main problem of using sigmoid as activation function in an artificial neural network (ANN)?<br>
**The sigmoid function's output is between [0,1]. However, looking to the sigmoid curve, when our z value is lower than say -2 or higher than 2, the derivatives will be all 0. This will make it harder to backpropagate because the gradients will be 0 or very close to 0. This will lead to a problem called vanishing gradients. Another problem with the sigmoid activation is that, when dealing with neural networks we prefer activation functions that are zero centered. This will make unsmooth learning curve (loss vs iteration).**

>d) What does it mean to overfit your data model?<br>
**When training our model on the data, if all the variance or more than enough variance from our training set comes to the learning and if our model becomes predicting nearly perfectly to a training set but fails dramatically when it encounters a dataset it never saw, we call this overfitting to the data. This problem is caused by small data set, too complex models or models that are too sensitive to the noise in data.**

  >e) Your input image size is 3x64x64. If you apply 3x3 convolution with input_channel=3, output_channel=6, padding=0, stride=2, what would be the size of the output?<br>
**Output size = (image volume - filter size + 2 x padding) / stride + 1

```python
class SceneDataset(Dataset):
    def __init__(self, image_paths, transform=False):
        self.image_paths = image_paths
        self.transform = transform
        
    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_filepath = self.image_paths[idx]
        image =  np.array(Image.open(image_filepath))
        import torchvision.transforms 
        transform = torchvision.transforms.Compose([
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225],)
        ])
        image = transform(image)
        label = image_filepath.split('/')[-2]
        label = class_to_idx[label]
        return image, label
```

###Model Implementation [7 pts]

Now implement your CNN. ConvNet class will represent your convolutional neural network. Implement 3 layers of convolution: 
<ul>
    <li>(1) 4 filters with size of 3 x 3 with stride 1 and padding 1, (2) ReLU </li>
    <li>(3) 8 filters with size of 3 x 3 with stride 1 and padding 1, (4) ReLU and (5) MaxPool 2 x 2 </li>
    <li>(6) 16 filters with size of 3 x 3 with stride 1 and padding 1, (7) ReLU and (8) MaxPool 2 x 2 </li> 
</ul>

As the classifier layer, you need to add only one linear layer at the end of the network. You need to choose the appropriate input and output neuron sizes and the activation function for the dense layer.

```python
class ConvNet(nn.Module):
    def __init__(self, **kwargs): 
        super(ConvNet, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=4, stride = 1, padding=1, kernel_size=(3,3))
        self.conv2 = nn.Conv2d(in_channels=4, out_channels=8, stride = 1, padding=1, kernel_size=(3,3))
        self.conv3 = nn.Conv2d(in_channels=8, out_channels=16, stride = 1, padding=1, kernel_size=(3,3))
        self.maxpool = nn.MaxPool2d(2)
        self.flatten = nn.Flatten(start_dim=1, end_dim=- 1)
        self.linear = nn.Linear(7744, 4)
        self.softmax = nn.LogSoftmax(dim=1)
    def forward(self, X): 
        X = F.relu(self.conv1(X))
        X = F.relu(self.conv2(X))
        X = self.maxpool(X)    
        X = F.relu(self.conv3(X))
        X = self.maxpool(X) 
        X = self.flatten(X)
        X = self.linear(X)
        X = self.softmax(X)
        return X
```
